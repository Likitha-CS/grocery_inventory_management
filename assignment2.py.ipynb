```python
# Define an empty dictionary to store the inventory
inventory = {}

# Define a function to add new items to the inventory
def add_item():
    item_name = input("Enter the name of the item: ")
    item_quantity = int(input("Enter the quantity of the item: "))
    item_price = float(input("Enter the price of the item: "))
    inventory[item_name] = {"quantity": item_quantity, "price": item_price}
    print(f"{item_name} has been added to the inventory.")

# Define a function to update existing items' quantities
def update_quantity():
    item_name = input("Enter the name of the item: ")
    if item_name in inventory:
        new_quantity = int(input("Enter the new quantity of the item: "))
        inventory[item_name]["quantity"] = new_quantity
        print(f"{item_name}'s quantity has been updated to {new_quantity}.")
    else:
        print(f"{item_name} is not in the inventory.")

# Define a function to view the current inventory
def view_inventory():
    print("Current Inventory:")
    for item_name, item_info in inventory.items():
        print(f"{item_name}: {item_info['quantity']} x ${item_info['price']}")

# Define a function to remove items from the inventory
def remove_item():
    item_name = input("Enter the name of the item: ")
    if item_name in inventory:
        del inventory[item_name]
        print(f"{item_name} has been removed from the inventory.")
    else:
        print(f"{item_name} is not in the inventory.")

# Define a function to display the menu and handle user input
def main_menu():
    while True:
        print("\nMenu:")
        print("1. Add new item to inventory")
        print("2. Update existing item's quantity")
        print("3. View current inventory")
        print("4. Remove item from inventory")
        print("5. Exit")
        choice = input("Enter your choice: ")
        if choice == "1":
            add_item()
        elif choice == "2":
            update_quantity()
        elif choice == "3":
            view_inventory()
        elif choice == "4":
            remove_item()
        elif choice == "5":
            break
        else:
            print("Invalid choice. Please try again.")

# Call the main menu function to start the program
main_menu()
```
